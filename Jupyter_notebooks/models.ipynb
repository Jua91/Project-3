```python
import pandas as pd

df = pd.read_csv("../Resources/student-mat.csv",sep=";",header=0)
```

```python
df.head()
```

```python
grade = []

for x in df['G3']:
    if x > 15:
        grade.append(5)
    elif x > 10:
        grade.append(4)
    elif x >5:
        grade.append(3)
    else:
        grade.append(2)
        
print(grade)
```

```python
df['grade'] = grade
```

```python
columns = df.keys()
columns
```

```python
len(df)
```

```python
new_df = df.drop(columns=['G1','G2','G3'])
new_df
```

```python
import numpy as np
X = df.drop(columns = 'grade')
y = df['grade'].values.reshape(-1,1)
```

```python
X.head()
```

```python
X_columns = ['school','sex','address','famsize','Pstatus','Mjob','Fjob','reason','guardian','schoolsup','famsup','paid','activities','nursery','higher','internet','romantic']
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()

for column in X_columns:
    label_encoder.fit(X[column])
    # label_encoder.classes_
    X[column] = label_encoder.transform(X[column])
```

```python
X
```

```python
X[X_columns] = X[X_columns].apply(LabelEncoder().fit_transform)
```

## Logistic Regression

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

from sklearn import tree

clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
clf.score(X_test, y_test)
```

```python
feature_names = X.columns
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=200)
rf = rf.fit(X_train, y_train)
sorted(zip(rf.feature_importances_, feature_names), reverse=True)
```

```python
rf.score(X_test, y_test)
```
