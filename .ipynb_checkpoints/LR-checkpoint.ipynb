```python
import matplotlib.pyplot as plt
import pandas as pd
import os
import numpy as np
import tensorflow
```

```python
 df = pd.read_csv('student-mat-por_combinedData.csv')
```

```python
grade = []
for x in df["G3"]:

    if x > 15:
        grade.append(4)
    elif x >= 10:
        grade.append(3)
    elif x >= 5:
        grade.append(2)
    else:
        grade.append(1)

df["grade"] = grade
df.head()
```

```python
new_df = df.drop(columns = ["G2","G1","G3"])
new_df
```

```python
data = new_df.values
X = new_df.drop(columns =["grade"])
y = new_df['grade']
y
```

```python
cat_columns = ["school","absences","age","health","Walc","goout","subject"]
columns = df.columns
print(columns)
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
label_encoder = LabelEncoder()
```

```python
X= X[cat_columns]
##X= X.apply(LabelEncoder().fit_transform)
X[["school","subject"]] = X[["school","subject"]].apply(LabelEncoder().fit_transform)
X
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler
X_scaler = StandardScaler().fit(X_train)
```

```python
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# import sklearn
# from tensorflow.keras.utils import to_categorical
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(random_state=42).fit(X_train,y_train)
lr.score(X, y)
```

```python
predictions = lr.predict(X_test)
print(predictions[:5])
print(y_test[:5])
```

```python
from sklearn.feature_selection import f_regression
```

```python
f_regression(X, y, center=True)
```

```python
# Create the GridSearchCV model
from sklearn.model_selection import GridSearchCV

param_grid = {'C': [5, 10, 50], 'max_iter': [200,500,1000],'penalty' : ['l1', 'l2'],}
grid_LR = GridSearchCV(lr, param_grid, verbose=3)
```

```python
# Train the model with GridSearch
grid_LR.fit(X_train_scaled, y_train)
```

```python
print(grid_LR.best_params_)
print(grid_LR.best_score_)
```

```python
# Make predictions with the hypertuned model
new_predictions = grid_LR.predict(X_test_scaled)
print(new_predictions[:5])
print(y_test[:5])
```
